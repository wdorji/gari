```python
import requests
from bs4 import BeautifulSoup

acc_story_urls = set()
for i in range(1,33):
    
    URL = f"http://www.bbs.bt/news/?s=road+accident&paged={i}"
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="tdi_97")
    job_elements = results.find_all("a")

    for l in job_elements:
        if "accident" in l["title"].lower():
            acc_story_urls.add(l["href"])
    print(f"Page {i} is done!")
acc_story_urls = sorted(list(acc_story_urls))
```

```python
import spacy
import pandas as pd
nlp = spacy.load("en_core_web_sm")

df_dict = {}
df_dict["published_time"] = []
df_dict["title"] = []
df_dict["text"] = []
df_dict["time"] = []
df_dict["location"] = []
df_dict["height"] = []



for u in acc_story_urls:

    cur_txt = ""
    cur_res = BeautifulSoup(requests.get(u).content, "html.parser").find(id="tdi_86")
    title = cur_res.find("h1").text
    pub_time = cur_res.find("time").text

    cur_res = cur_res.find_all("p")
    for text_ele in cur_res:
        cur_txt += text_ele.text.strip()

    doc = nlp(cur_txt)
    time = [ent.text for ent in doc.ents if ent.label_ == "TIME"]

    try:
        location = cur_txt.split("at")[1].split(".")[0]
    except:
        location = "no location info"
    
    height = "no metre height"
    if "metre" in cur_txt.lower():
        height = cur_txt.split("metre")[0].split()[-1]
    if "meter" in cur_txt.lower():
        height = cur_txt.split("metre")[0].split()[-1]

    df_dict["text"].append(cur_txt)
    df_dict["time"].append(time)
    df_dict["location"].append(location)
    df_dict["height"].append(height)
    df_dict["published_time"].append(pub_time)
    df_dict["title"].append(title)



pd.DataFrame.from_dict(df_dict)
```
